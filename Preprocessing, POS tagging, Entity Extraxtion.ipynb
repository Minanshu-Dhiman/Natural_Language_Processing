We will use requests library to request the HTML text from (Early Life Of Ratan Tata and History Of Netfilx) Wikipedia so that it can be used further for information extraction by using BesautifulSoup library.

```python
import requests
from bs4 import BeautifulSoup
```

```python
headers = {"User-Agent": "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36(KHTML, like Gecko) Chrome/88.0.4324.96 Safari/537.36"}
url_Ratan_Tata_wiki = "https://en.wikipedia.org/wiki/Ratan_Tata"
url_Netflix_wiki = "https://en.wikipedia.org/wiki/Netflix"
a = requests.get(url_Ratan_Tata_wiki, headers=headers)
html_doc1 = a.content
b = requests.get(url_Netflix_wiki, headers=headers)
html_doc2 = b.content
soup_Ratan = BeautifulSoup(html_doc1, 'html.parser')
soup_Netflix = BeautifulSoup(html_doc2, 'html.parser')
Ratan_Tata_HTML = soup_Ratan.prettify()
print(Ratan_Tata_HTML,"\n")
Netflix_HTML = soup_Netflix.prettify()
print(Netflix_HTML)
```

In the HTML text of wiki page of Mr.Ratan Tata, we have to extract the paragraph containing information about his early life, that text is present in **P** tag 5 and 6 inside a div of class **"mw-content-ltr mw-parser-output"**, so we will extract that into a string, using BeautifulSoup.

```python
div_Ratan = soup_Ratan.find("div", class_="mw-content-ltr mw-parser-output")
p_tags = div_Ratan.find_all("p")
Early_life = ""
for i in range(5, 7):
    Early_life += p_tags[i].get_text()
print(Early_life)
```

In the HTML text of wiki page of Netflix we have to extract history of Netflix, which is present present in **P** tag 5 and 11 inside a div of class **"mw-content-ltr mw-parser-output"**, so we will extract that into a string, using BeautifulSoup.

```python
div_Netflix= soup_Netflix.find("div", class_="mw-content-ltr mw-parser-output")
p_tags = div_Netflix.find_all("p")
History = ""
for i in range(5, 12):
    History += p_tags[i].get_text()
print(History)
```

Now our task is to perform **Stemming and POS tagging**. Spacy doesn't support **Stemming**, it only support **lematization**. So for that matter we would have to use **nltk** for **Stemming**. For **Stemming** we would firstly have to **tokenize** our text so that we are able to stemm it. The **Function tokenize** tokenizes the string into words.

```python
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
def tokenize(paragraph):
    words = word_tokenize(paragraph)
    return words
```

Now we will tokenize both the data sets that we have(**Early_life, and History**), and perform **Stemming** on both.

```python
words_Ratan = tokenize(Early_life)
print(words_Ratan)
for word in words_Ratan:
    print(word, "| Base Word: ", stemmer.stem(word))
```

```python
words_Netflix = tokenize(History)
print(words_Netflix)
for word in words_Netflix:
    print(word, "| Base Word: ", stemmer.stem(word))
```

Now for **POS Tagging** we are going to use **spacy**, we have the data extracted into strings we just have to use **spacy** to perforn **POS Tagging**

```python
import spacy
nlp = spacy.load("en_core_web_sm")
```

**POS Tagging** of **Early Life** of Mr. Ratan Tata

```python
doc_Ratan= nlp(Early_life)
for token in doc_Ratan:
    print(token,  " | ", spacy.explain(token.pos_))
```

**POS Tagging** of **History** of **Netflix**

```python
doc_Netflix = nlp(Early_life)
for token in doc_Netflix:
    print(token,  " | ", spacy.explain(token.pos_))
```

**Lemmatization** of **Early life** of Mr. Ratan Tata.

```python
for token in doc_Ratan:
    print(token, " |  Lemma: ", token.lemma_)
```

**Lemmatization** of **History** of Netflix.

```python
for token in doc_Netflix:
    print(token, " |  Lemma: ", token.lemma_)
```

For **Entity extraction** we will use **displacy**, a function in spacy that represents entities **Graphical** embeded into the sentences.

```python
from spacy import displacy
```

**Entity Extraction** form **Early life** of Mr. Ratan Tata

```python
displacy.render(doc_Ratan, style = "ent")
```

**Entity Extraction** form **History** of Netflix
